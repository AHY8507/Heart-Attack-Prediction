```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("/kaggle/input/heart-attack-analysis-prediction-dataset/heart.csv")
print(df.shape)
df.head()
```

```python
df.dtypes
```

```python
x = df.drop(columns="output").values
y = df.output.values
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler().fit(x)
x = scaler.transform(x)
x
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score

max_accure = -1e3
train_x, test_x, train_y, test_y = train_test_split(x, y, test_size = 0.2, random_state = 42)
print(train_x.shape, test_x.shape)
print(train_y.shape, test_y.shape)
```

<h3>Support Vector Machine

```python
from sklearn.svm import SVC

kernels = ["linear", "rbf", "sigmoid", "poly"]

for kernel in kernels:
    clf = SVC(kernel = kernel)
    clf.fit(train_x, train_y)
    
    y_hat = clf.predict(test_x)
    max_accure = max(max_accure, accuracy_score(test_y, y_hat))
    
    print(kernel)
    print(classification_report(test_y, y_hat), end="_______________________________________________________\n\n\n")
```

<h3>Logistic Regression

```python
from sklearn.linear_model import LogisticRegression

solvers = ["lbfgs", "liblinear", "newton-cg", "sag", "saga"]

for solver in solvers:
    clf = LogisticRegression(solver = solver)
    clf.fit(train_x, train_y)
    
    y_hat = clf.predict(test_x)
    max_accure = max(max_accure, accuracy_score(test_y, y_hat))
    
    print(solver)
    print(classification_report(test_y, y_hat), end="_______________________________________________________\n\n\n")
```

<h3>Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier

clf = DecisionTreeClassifier().fit(train_x, train_y)

y_hat = clf.predict(test_x)
max_accure = max(max_accure, accuracy_score(test_y, y_hat))

print(classification_report(test_y, y_hat))
```

```python
from sklearn.neighbors import KNeighborsClassifier

for n in range(1, 50):
    clf = KNeighborsClassifier(n_neighbors = n).fit(train_x, train_y)
    
    y_hat = clf.predict(test_x)
    max_accure = max(max_accure, accuracy_score(test_y, y_hat))

    print(n)
    print(classification_report(test_y, y_hat), end="_______________________________________________________\n\n\n")
    
```

<h3>Maximum Accuracy

```python
print('Maximum Accuracy:', max_accure)
```
